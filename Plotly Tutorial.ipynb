```python
def profit(buy, sell, quantity):
    cost = buy * quantity
    shmoney = sell * quantity
    profit = shmoney - cost
    return profit
```

```python
stonks = 1000
step = 5
cur = 200
while cur < stonks:
    print(f'Sell for ${cur} get ${profit(200, cur, 1)}')
    cur += step
```
